# Setting Up Continuous Peak Fit - Windows 10

This document details the installation procedure needed in order to successfully run the Continuous Peak Fit (`cpf`) Python package on a Windows 10 OS.

## Prerequisites
1. `ImageD11` (v1.9.8)  
    CPF borrows a lot of functions from this package.
2. Visual Studio Build Tools  
    The C++ development tools from this package are needed in order to successfully install ImageD11. The installer can be found at:  
    https://aka.ms/vs/17/release/vs_BuildTools.exe
3. `Python` (3.8)  
    ImageD11 seems to be incompatible with the newer versions of Python. As of now, Python 3.8 seems to be the latest version that is compatible with ImageD11.

## Installation
1. Create a new Python 3.8 environment (optional).  
    * Create new environment using `conda`:  
    https://conda.io/projects/conda/en/latest/user-guide/tasks/manage-environments.html#creating-an-environment-with-commands
2. Use `pip` to install `ImageD11` (1.9.8) in this version.

## Manual Addition and Updating of Packages
Having successfully installed `ImageD11`, the followings additional packages will have to be installed manually (along with their dependencies) in order to allow the `cpf` functions to work correctly:
1. `pathos` (0.3.1)
2. `hdf5plugin`(4.2.0)
3. `lmfit` (1.2.2)
4. `moviepy` (1.0.3)
5. `h5py` (3.10.0)
6. `opencv-python` (4.8.1.78)
7. `pyfai-2023.9.0`
8. `numexpr` (2.8.3) - Downgrade; the latest version 2.8.6 seems to generate unrecognised characters
9. `pandas` (2.0.3)

Optional packages to allow the environment to be used in Jupyter Notebook and Spyder
1. `spyder-kernels` (Spyder)  
    Connecting your new Python environment to Spyder:  
    https://github.com/spyder-ide/spyder/wiki/Working-with-packages-and-environments-in-Spyder#working-with-other-environments-and-python-installations
2. `ipykernel` (Jupyter Notebook)  
    Setting up `ipykernel` in the environments you want to use:  
    https://ipython.readthedocs.io/en/stable/install/kernel_install.html#kernels-for-different-environments
